***BARPLOT***


---



***Bar Chart: Average Cost per Sub-Type***
<br>*This bar plot shows the average ad cost (in USD) for each ad subcategory (like Finance, Games, Social Media, etc.).*

```r


# Calculate average cost per subcategory
avg_cost_data <- processed_data %>%
  group_by(sub_type) %>%
  summarise(avg_cost = mean(cost, na.rm = TRUE))


ggplot(avg_cost_data, aes(x = sub_type, y = avg_cost, fill = sub_type)) +
  geom_bar(stat = "identity") +
  scale_fill_brewer(palette = "Set2") +
  labs(
    title = "Average Ad Cost by Subcategory",
    x = "Ad Subcategory",
    y = "Average Cost (USD)"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1),
        legend.position = "none")
```

*E-commerce ads have the highest average cost, slightly above $13.*

*Shopping and Finance follow closely, averaging around USD 12 –  USD 13 and USD 10,respectively.*

*Entertainment and Games fall in the mid-range, with average costs between USD 7 – USD 9.*

*Social Media and Travel have the lowest average costs, with Travel being the cheapest (around $4).*


***Inference:***

1. *E-commerce and Shopping categories likely involve high-conversion potential,
which justifies higher ad spending. Advertisers may be willing to invest more due to strong ROI opportunities.*

2. *Finance, though traditionally high-cost, appears slightly lower than E-commerce, suggesting some efficiency or more precise targeting.*

3. *Low-cost categories like Travel and Social Media could be ideal for broad reach campaigns on a limited budget.*

4. *Advertisers should prioritize high-cost categories for premium products or lead generation, and explore lower-cost sectors for awareness campaigns or experimental ad formats.*

*Line Plot: Distribution of Ad Costs with Density

*This chart is a histogram with a density line overlay that shows the distribution of ad costs.*

*The bars show the count of ads that fall into different cost ranges, with each bin covering $5 intervals.*

*The line is a density curve scaled to match the histogram’s count scale (multiplying the density by the binwidth), which provides a smoothed view of how ad costs are distributed.*

There are some smaller peaks as well, coming around 7000 ads for an ad cost of 3 dollars and 10 dollars, respectively. ***PIE CHART***

<br>

*Pie Chart: Average Cost per Sub-Type*
<br><br>
*This pie chart shows the distribution of ads across different sub-types. Each slice of the pie represents a different ad sub-type (from the sub_type column in the data).*

*The size of each slice is proportional to the number of ads that fall into that sub-type.*

```r
# Count the number of ads in each sub_type
subtype_counts <- processed_data %>%
  group_by(sub_type) %>%
  summarise(count = n()) %>%
  mutate(percentage = count / sum(count) * 100,
    label = paste0(round(percentage, 1), "%"))
# Make a pie chart
ggplot(subtype_counts, aes(x = "", y = count, fill = sub_type)) +
  geom_col(width = 1, color = "white") +
  coord_polar(theta = "y") +
  geom_text(aes(label = label),
            position = position_stack(vjust = 0.45),size = 4)+
  scale_fill_brewer(palette = "Set3")+
  labs(title = "Ad Distribution by Sub-Type") +
  theme_void() +
  theme(legend.title = element_blank())
```

*Games and Finance industries are major drivers of ad volume.
Shopping and Social Media both tie for the industry with the least amount of ad distribution volume.*

***Inferences***:<BR>
*Games and Finance industries are potentially high due to high user engagement or monetization potential.

***RECOMMENDATIONS***

---


Add Additional Features to Improve Insight & Predictive Accuracy

While our current dataset provides valuable metrics like ad cost, type, rating, sub_type, and predicted_ctr, these attributes alone limit our ability to understand the full context of ad performance. To enhance insight, we recommend collecting additional contextual and behavioral features.

    These features will:


1.   Improve model explainability (why CTR is high/low),
2.   Boost prediction accuracy
3.   And support more actionable business decisions.



**Recommended Features & Their Impact:**

1. **User Demographics** (Age, Gender, Location)
Why it matters: Ads often perform differently across user groups. A 25-year-old may engage more with gaming ads, while a 45-year-old may prefer finance ads.

 **Value added**: Enables segmentation, targeting, and personalization.

 **Result**: Allows models to predict CTR/cost for specific audience segments → better ROI through audience targeting.

2. **Device Type** (Mobile, Desktop, Tablet)
Why it matters: User behavior changes across devices, mobile users may scroll faster, desktop users may engage longer.

 **Value added**: Helps compare CTR patterns and optimize creatives/formats for each platform.

 **Result**: Improves ad delivery strategy; mobile-optimized ads shown to mobile users → higher engagement.

3. **Ad Placement** (Homepage, Sidebar, Footer, In-feed)
Why it matters: Visibility impacts CTR — ads at the top of a page typically perform better than those in the footer.

 **Value added**: Identifies high-performing placements.

 **Result**: Lets advertisers pay a premium for better placements and improve ROI on each ad slot.

4. Time Features (Day of Week, Hour of Day)
Why it matters: User engagement fluctuates — e.g., shopping ads may perform better on weekends, job ads during weekdays.

 **Value added**: Reveals peak times for each ad type or audience.

 **Result**: Improves scheduling; ads shown at high-CTR windows → better performance with the same budget.

5. **Ad Format / Media Type** (Image, Video, Carousel)
Why it matters: Different creatives have different effectiveness — videos may have higher CTR but also higher cost.

 **Value added**: Enables format-level A/B testing and creative optimization.

 **Result**: Tailors ad design strategy to maximize engagement per format type.
